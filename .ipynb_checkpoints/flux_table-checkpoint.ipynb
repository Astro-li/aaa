```python
magu_tar = []
magz_tar = []
magu_err_tar = []
magz_err_tar = []
idx_no = []
d_theta = []
for i,ra in enumerate(pecat[:,1]):
    idx = where((ra1244-ra)**2+(dec1244-pecat[i,2])**2==((ra1244-ra)**2+(dec1244-pecat[i,2])**2).min())[0][0]
    if (sqrt((ra1244-ra)**2+(dec1244-pecat[i,2])**2)).min()>0.5/3600:
        print((sqrt((ra1244-ra)**2+(dec1244-pecat[i,2])**2)).min())
        magu_tar.append(99)
        magz_tar.append(99)
        magu_err_tar.append(99)
        magz_err_tar.append(99)
        idx_no.append(i)
        d_theta.append((sqrt((ra1244-ra)**2+(dec1244-pecat[i,2])**2)).min())
    else:
        magu_tar.append(magu[idx])
        magz_tar.append(magz[idx])
        magu_err_tar.append(magu_err[idx])
        magz_err_tar.append(magz_err[idx])
```

```python
idx
```

```python
magk_tar = []
magk_err_tar = []
idx_no_k = []
d_theta_k = []
for i,ra in enumerate(pecat[:,1]):
    idx = where((k1244[:,1]-ra)**2+(k1244[:,2]-pecat[i,2])**2==((k1244[:,1]-ra)**2+(k1244[:,2]-pecat[i,2])**2).min())[0][0]
    if (sqrt((k1244[:,1]-ra)**2+(k1244[:,2]-pecat[i,2])**2)).min()>0.5/3600:
        print(sqrt(((k1244[:,1]-ra)**2+(k1244[:,2]-pecat[i,2])**2)).min())
        magk_tar.append(99)
        magk_err_tar.append(99)
        idx_no_k.append(i)
        d_theta_k.append((sqrt((ra1244-ra)**2+(dec1244-pecat[i,2])**2)).min())
    else:
        magk_tar.append(k1244[idx,4])
        magk_err_tar.append(k1244[idx,5])
```
